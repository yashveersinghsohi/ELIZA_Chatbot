### Defining Reflections
**Reflections** is a key-value pair dictionary that matches the pronouns used by the user to the ones that are to be used by our chat-bot.

```python
Reflections = {
"am"   : "are",
"was"  : "were",
"i"    : "you",
"i'd"  : "you would",
"i've"  : "you have",
"i'll"  : "you will",
"my"  : "your",
"are"  : "am",
"you've": "I have",
"you'll": "I will",
"your"  : "my",
"yours"  : "mine",
"you"  : "me",
"me"  : "you"
}
```

### Defining function to map Reflections.
The **reflect** function is used to return a appropriate pronoun that will change the perspective of the statement - from first hand(user's point of view) to second hand(chat-bot's point of view). The function will return a different value from the input given only if the input matches the list of user pronouns mentioned in the **Reflections** dictionary. If no match is found for the input in the **Reflections** dictionary, then the input is returned as it is.

```python
def reflect(user_key):
    tokens = user_key.split()
    for i, token in enumerate(tokens):
        if token in Reflections:
            tokens[i] = Reflections[token]
    return ' '.join(tokens)
```

### Defining the function to compute the response from the chat-bot
This function, matches the user's input to a set of patterns defined in **pattern_responses_list** and returns a random response from the corresponding list of responses along with an ideal reflection, using the **reflect** function. In case of no match a response from the **default_responses** list is printed.

```python
def bot_response(user_response):
    flag = 0
    user_response = user_response.lower()
    for pattern, responses in pattern_responses_list:
        match = re.match(pattern.lower(), user_response)
        if match:
            flag = 1
            response = random.choice(responses)
            return response.format(*[reflect(g) for g in match.groups()])
    if flag == 0:
        return random.choice(default_responses)
```

### Defining the main function
An introductory message is printed in the begining and then the user's input is taken till they enter the keyword **bye**. For each input, the **bot_response** function is called with appropriate dealy to mimic a real chat.

```python
print("Hello. I'm ELIZA. Let's start our session")
user_response = str()
while(user_response!="bye"):
    print()
    time.sleep(0.2)
    user_response = input("YOU: ")
    user_response = user_response.lower()
    if(user_response!='bye'):
        print("BOT: "+bot_response(user_response))
    else:
        time.sleep(0.2)
        print("BOT: Bye! take care..")
```
